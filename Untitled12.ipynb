```python
import whisperx

device = "cuda"
audio_file = "audio.mp3"
batch_size = 16
compute_type = "float16"

transcribe = whisperx.load_model("large-v2", device, compute_type = compute_type)
```

```python
audio = whisperx.load_audio(r"/content/converted.mp3")
transcript = transcribe.transcribe(audio, batch_size= batch_size)
```

```python
model_a, metadata = whisperx.load_align_model(language_code=transcript["language"], device=device)
result = whisperx.align(transcript["segments"], model_a, metadata, audio, device, return_char_alignments=False)
```
